- names can not contain any of these symbols:

```
# Casting an into a float
my_integer = 3
my_integer+=0.1
type(my_integer)
```

### Strings (str): Sequences of characters, enclosed in single or double quotes.

```
my_string = "Hello, World!"
type(my_string)
```

```
# Strings are sequences, so we can do sequence operations:
palindrome = "racecar"
is_palindrome = palindrome == palindrome[::-1]
palindrome[::-1]
```

```
# Multiplying strings:
cheer = "Hip " * 3 + "Hooray!"
cheer
```

```
# Fun with String Slicing

sentence = "Python is fun!"

# Basic slicing
first_five_chars = sentence[:5]

# Slicing with negative indices
last_five_chars = sentence[-5:]

# Slicing with steps
alternate_chars = sentence[::2]
reverse_string = sentence[::-1]

# Slicing beyond the string's length doesn't raise an error
beyond_length = sentence[5:100]

# A slice of a string can be used to form a new word
word_play = sentence[7:10] + sentence[1:3]

# Let's see the sliced and diced results!
first_five_chars, last_five_chars, alternate_chars, reverse_string, beyond_length, word_play
```

### Tuples: Ordered collections like lists, but immutable.

```
# Fun with Tuples

# Creating a basic tuple
animals = ("cat", "dog", "elephant")

# Unlike lists, tuples are immutable. Let's see what happens when we try to change one.
try:
    animals[1] = "dolphin"
except TypeError as e:
    error_message = str(e)

animals, error_message
```

```
# Tuples can be concatenated
more_animals = animals + ("fish", "giraffe")

# Nesting tuples
zoo = (animals, ("lion", "zebra", "kangaroo"))

zoo
```

```
# Tuple unpacking!
cat, dog, elephant = animals
unpacked_animals = f"Unpacked: {cat}, {dog}, and {elephant}"

unpacked_animals
```

```
# Fun with Tuple Slicing

# A journey through time and space with Doctor Who's TARDIS
tardis_journey = ("1963", "Medieval England", "Victorian London", "Mars", "End of Time", "Gallifrey")

# Let's take a short trip to the first three destinations
short_trip = tardis_journey[:3]

short_trip
```

```
# Oops! The TARDIS overshot. We're now somewhere at the end of the universe!
end_of_universe = tardis_journey[-2:]

end_of_universe
```

```
# Timey-wimey stuff! Let's jump through every alternate location
alternate_jumps = tardis_journey[::2]
alternate_jumps
```

```
# Let's take a detour and reverse our journey
reverse_journey = tardis_journey[::-1]
reverse_journey
```

```
# A slice from another dimension (slicing beyond the tuple's length doesn't cause an error)
beyond_time = tardis_journey[4:100]
beyond_time
```

### Dictionaries (dict): Unordered collections of key-value pairs.

```
# Basic dictionary
favorite_fruits = {
    "Alice": "Apple",
    "Bob": "Banana",
    "Charlie": "Cherry"
}

# Accessing a key that doesn't exist... oops!
try:
    daves_fruit = favorite_fruits["Dave"]
except KeyError as e:
    error_message = f"KeyError: {str(e)}"

# Using the `get` method to provide a default value for missing keys
daves_fruit = favorite_fruits.get("Dave", "Unknown fruit")

# Dictionary comprehensions! Let's reverse our dictionary
reversed_fruits = {value: key for key, value in favorite_fruits.items()}

error_message, daves_fruit, reversed_fruits
```

```
# Modifying the favorite_fruits dictionary
favorite_fruits["Dave"] = "Dragonfruit"

favorite_fruits
```

```
# Keys and values
fruit_keys = list(favorite_fruits.keys())
fruit_values = list(favorite_fruits.values())

fruit_keys, fruit_values
```

### Sets: Unordered collections of unique values.

Note: To create an empty set in Python, you need to use set(). Using {} would create an empty dictionary, not a set.

Sets in mathematics refer to a collection of distinct (unique) objects, considered as an entity itself. Set theory is the branch of mathematical logic that studies sets, their operations, and the properties and relations of these operations.

$ S_1 = \{1, 2, 3, 4, 5\} $
$ S_2 = \{\text{apple}, \text{banana}, \text{cherry}\} $

An xample of Maths sets operations is union:

$ S_1 \cup S_2  = \{1, 2, 3, 4, 5, \text{apple}, \text{banana}, \text{cherry}\} $

```
# Using curly braces
fruits = {"apple", "banana", "cherry"}

# Using the set() constructor
vegetables = set(["broccoli", "carrot", "spinach"])
```

```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

union_set = A | B

union_set
```

```
intersection_set = A & B
intersection_set
```

```
difference_set = A - B
difference_set
```

### The `None` Type

The `None` Type in Python represents the absence of a value or a null value. The only value of this type is the None value. It's often used to signify the absence of a value in functions, default arguments, and data structures, among other places.

```
# Different ways you might encounter None

# A function that doesn't return anything, actually returns None
def do_nothing():
    pass

result = do_nothing()

# Variables can be assigned the value of None
empty_variable = None

# Dictionary values or list elements can be None
sample_dict = {"key": None}
sample_list = [None, 1, 2]

# Check type
none_type = type(None)

result, empty_variable, sample_dict, sample_list, none_type
```

- parameters: (Optional) Values you can pass into the function.
- docstring: (Optional) A brief description of what the function does.
- return: (Optional) This keyword is used to send a result back to the caller.

```
# function definition

def function_name(parameters):
    """docstring"""
    # function body
    return value
```

```
# function definition

def greet(name):
    """This function greets the person passed in as a parameter."""
    return f"Hello, {name}!"
```

```
# function call

message = greet('Alice')
print(message)  # Output: Hello, Alice!
```
